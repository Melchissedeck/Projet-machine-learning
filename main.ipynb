#2. Normalisation de la Donnée


##Justification de l’intérêt de normaliser les données (notamment pour des méthodes sensibles aux échelles, comme KNN et PCA).

```
Certaines variables peuvent avoir des ordres de grandeur très différents. Exemple :

Taille (cm) : entre 150 et 200.
Poids (kg) : entre 50 et 100.
Salaire (€) : entre 2 000 et 10 000.
Dans ce cas, les modèles qui utilisent des distances entre les points donneront plus d’importance aux variables avec des valeurs élevées, ce qui biaisera les résultats.

Exemples d’algorithmes impactés :

K-Nearest Neighbors (KNN)
Méthodes basées sur la distance (K-Means, SVM avec certains noyaux, etc.)
Réduction de dimension avec PCA
```

```
- Cas particulier de KNN (K-Nearest Neighbors)
KNN classe un point en fonction de ses voisins les plus proches. La distance entre les points est donc un élément clé.

Si une variable a une plus grande plage de valeurs que les autres, elle domine la distance. Par exemple :

Une différence de 10 cm en taille paraît faible comparée à une différence de 5 000 € en salaire.
Résultat : le modèle KNN privilégiera le salaire pour calculer la proximité, ignorant presque la taille.
```

```
Cas particulier de PCA (Analyse en Composantes Principales)
PCA est une méthode de réduction de dimension qui cherche les directions où la variance est maximale.

⚠️ Problème sans normalisation :
Les variables avec de grandes valeurs (ex. salaire) auront une variance plus élevée, ce qui fausse l’analyse :

Les premières composantes principales seront dominées par ces variables.
On risque de perdre des informations importantes venant d'autres variables.
```

#3. Réduction de Dimensionnalité avec la PCA

##Présentation des principes de la PCA

```
- Principe général
La PCA (Principal Component Analysis) est une technique de réduction de dimension qui cherche de nouvelles variables (appelées composantes principales) qui expliquent au mieux la variance du jeu de données.
Ces nouvelles composantes sont des combinaisons linéaires des variables d’origine.
Elles sont orthogonales entre elles (indépendantes).
La première composante principale (PC1) est celle qui explique la plus grande partie de la variance des données, la deuxième (PC2) explique la plus grande partie de la variance résiduelle.

- Objectifs
Réduire la dimension : Passer par exemple de 20 variables à 2 (ou plus) sans trop perdre d’information.
Visualisation : En 2D ou 3D, il est plus aisé de visualiser la répartition des individus et de repérer d’éventuels clusters.

-Pré-requis : Standardisation
Il est vivement conseillé de standardiser les données (centrage-réduction) avant d’appliquer la PCA, car la PCA est sensible à l’échelle des variables.
```

## Réduction des données à 2 composantes principales pour faciliter la visualisation.

````python
from sklearn.decomposition import PCA  # On importe PCA pour la réduction de la dimensionnalité

pca_2D = PCA(n_components=2)  # On crée une instance de PCA pour réduire les données à 2 dimensions
X_pca_2D = pca_2D.fit_transform(X_scaled)  # On applique PCA et on transforme les données en 2 dimensions
````

## Réduction des données à 3 composantes principales

````python
from sklearn.decomposition import PCA  # On importe PCA de sklearn pour réduire la dimensionnalité des données

pca_3D = PCA(n_components=3)  # On crée une instance de PCA pour réduire à 3 dimensions
X_pca_3D = pca_3D.fit_transform(X_scaled)  # On applique PCA pour transformer les données en 3 dimensions
````

###b) Scree Plot

````python
import numpy as np  # Importation de numpy pour les calculs mathématiques

# Entraînement de la PCA sans limite sur le nombre de composantes
pca_full = PCA()  # PCA avec toutes les composantes
pca_full.fit(X_scaled)  # Apprentissage de la PCA sur les données normalisées

# Création du Scree Plot
plt.figure(figsize=(8,5))  # Définition de la taille de la figure
plt.plot(
    range(1, len(X.columns)+1),  # Nombre de composantes
    np.cumsum(pca_full.explained_variance_ratio_),  # Variance expliquée cumulée
    marker='o'  # Marqueurs sur le graphique
)
plt.title("Scree Plot : Variance expliquée cumulée")  # Titre du graphique
plt.xlabel("Nombre de composantes")  # Légende de l'axe des X
plt.ylabel("Variance expliquée cumulée")  # Légende de l'axe des Y
plt.grid(True)  # Ajout de la grille pour une meilleure lecture
plt.show()  # Affichage du graphique
````
